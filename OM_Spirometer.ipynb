## OM Spirometer

This Spirometer is a way to observe opionion leaders and influencers, those who promote values or memes relevant to the open making.

This service provides the followings

* Spirometer Categories
* Spirometer GUI
* List Opinon Learders
* List Opinon Learders in Context
* Overall Score of an Influencer

### Spirometer Categories

Returns the list of available category/memes spirals.

```python
import requests
import json

insightIP = 'http://178.62.229.16'
insightPort = '8484'
insightVersion = 'v1.1'

insightSetting = insightIP + ':' + insightPort + '/api/' + insightVersion 
api_key = "" #needs to be filled

request = '/spirometer?' + 'api_key=' + api_key

# send a request
res = requests.get(insightSetting + request)
```

```python
# show the response in JSON format
res.json()
```

### Spirometer GUI

Interactive Graphics of Spirometer in HTML.

```python
import requests

insightIP = 'http://178.62.229.16'
insightPort = '8484'
insightVersion = 'v1.1'

# parameters of spirometer gui is screen_name and api_key
screen_name = "instructables"

insightSetting = insightIP + ':' + insightPort + '/api/' + insightVersion 
api_key = "" #needs to be filled

request = '/spirometer/gui/api?' + 'screen_name=' + screen_name + '&' + 'api_key=' + api_key

# send a request
res = requests.get(insightSetting + request)
```

```python
# show a status code of the response
res.status_code
```

### List Opinion Leaders

Retrieving the list of overall opinion leaders on OpenMaking. The score at the moment is generated using a built-in dictionary, where a predetermined keywords and phrases are matched to categories. The app aims to maximize accuracy and retrieval at detecting the predetermined expresions within the texts that are generated by an influencer. Right now, it scans the tweets of the user as input texts in order to detect themes within them and map the themes to the categories.

```python
import requests
import json

insightIP = 'http://178.62.229.16'
insightPort = '8484'
insightVersion = 'v1.1'

insightSetting = insightIP + ':' + insightPort + '/api/' + insightVersion
api_key = "" #needs to be filled

request = '/spirometer/scoreboard?' + 'api_key=' + api_key

# send a request
res = requests.get(insightSetting + request)
```

```python
# show the response in JSON format
res.json()
```

### List Opinion Leaders in Context

Retrieveing the list of opinion leaders on a specific theme.

A specific board can be one of these: <br>
category:{"openness","sharing","innovation","sustainability", "collectiveness"}

```python
import requests
import json

insightIP = 'http://178.62.229.16'
insightPort = '8484'
insightVersion = 'v1.1'

# parameters of spirometer scoreboard is category and api_key
category = "sustainability"

insightSetting = insightIP + ':' + insightPort + '/api/' + insightVersion
api_key = "" #needs to be filled

request = '/spirometer/scoreboard?' + 'category =' + category

# send a request
res = requests.get(insightSetting + request)
```

```python
# show the response in JSON format
res.json()
```
